```python
import sqlite3
import numpy as np
import pandas as pd
```

```python
connection = sqlite3.connect("Books.db")
```

```python
def query(sql,data):
    global connection
    rows = connection.execute(sql,data)
    connection.commit()
    return rows
```

```python
sql= """
SELECT * FROM Book
WHERE Rate >= ? AND Name LIKE ?  
"""
data = [3,"%Mr%"]
s1   = query(sql,data)
```

```python
df = pd.DataFrame(s1,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT * FROM Book
WHERE Rate >= ? AND Price > ?  
"""
data = [3,20]
s2   = query(sql,data)
```

```python
df = pd.DataFrame(s2,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT * FROM Book
WHERE Rate != ?  
"""
data = [3]
s3   = query(sql,data)
```

```python
df = pd.DataFrame(s3,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT * FROM Book
WHERE Price BETWEEN ? AND ?  AND  Rate IN(?,?,?)  
"""
data = [10,40,3,4,1]
s4   = query(sql,data)
```

```python
df = pd.DataFrame(s4,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT * FROM Book
ORDER BY Price DESC LIMIT ?  
"""
data = [5]
s5   = query(sql,data)
```

```python
df = pd.DataFrame(s5,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT * FROM Book
ORDER BY Rate DESC , Price ASC LIMIT ? OFFSET ?
"""
data = [10,20]
s6   = query(sql,data)
```

```python
df = pd.DataFrame(s6,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
rows = connection.execute("INSERT INTO Category (Name) VALUES(?)",["DATABASE"])
lr   = rows.lastrowid
connection.commit()

data = [["SQL",444,5,6,lr],["DB",200,3,7,lr],["PLSQL",300,4,2,lr],["DATA",55.9,1,8,lr],["DBMS",100,4,2,lr]]
for i in data:
    connection.execute("INSERT INTO Book (Name,Price,Rate,Quantity,id_c) VALUES(?,?,?,?,?)",i)
    connection.commit()
```

```python
sql= """
UPDATE Book SET Rate=? 
WHERE Price < ?  
"""
data = [4,30]
s8   = query(sql,data)
```

```python
df = pd.DataFrame(s8,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
DELETE FROM Book
WHERE Price > ? AND Rate <= ?  
"""
data = [50,2]
s9   = query(sql,data)
```

```python
df = pd.DataFrame(s9,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```

```python
sql= """
SELECT COUNT(*) FROM Book
WHERE Name LIKE ? AND Price BETWEEN ? AND ? 
"""
data = ["Secret",10,25]
s10  = query(sql,data)
```

```python
df = pd.DataFrame(s10,columns=['Count'])
df.head()
```

```python
sql= """
SELECT MAX(Price) , MIN(Price) FROM Book
GROUP BY Rate = ? 
"""
data = [5]
s11  = query(sql,data)
```

```python
df = pd.DataFrame(s11,columns=['MAX','MIN'])
df.head()
```

```python
sql= """
SELECT AVG(Price) FROM Book
GROUP BY Rate = ? 
"""
data = [5]
s12  = query(sql,data)
```

```python
df = pd.DataFrame(s12,columns=['AVG'])
df.head()
```

```python
sql= """
SELECT SUM(Price) FROM Book
GROUP BY Rate = ? AND Price BETWEEN ? AND ?
"""
data = [2,10,40]
s13  = query(sql,data)
```

```python
df = pd.DataFrame(s13,columns=['SUM'])
df.head()
```

```python
sql= """
SELECT Book.Name,Book.Price,Book.Quantity,Category.Name FROM Book INNER JOIN Category
WHERE Book.id_c = Category.id
"""
data = []
s14  = query(sql,data)
```

```python
df = pd.DataFrame(s14,columns=['Name','Price','Quantity','Category'])
df.head()
```

```python
sql= """
SELECT Rate , COUNT(Name) FROM Book
WHERE Price BETWEEN ? AND ?
GROUP BY Rate
"""
data = [20,30]
s15  = query(sql,data)
```

```python
df = pd.DataFrame(s15,columns=['Rate','COUNT'])
df.head()
```

```python
sql= """
SELECT id_c , COUNT(Name) AS ACC FROM Book
GROUP BY id_c
HAVING ACC > ?
"""
data = [10]
s16  = query(sql,data)
```

```python
df = pd.DataFrame(s16,columns=['id_c','COUNT'])
df.head()
```

```python
sql= """
SELECT * FROM Book
WHERE id_c = (SELECT id FROM Category WHERE Name = ?)
"""
data = ["Music"]
s17  = query(sql,data)
```

```python
df = pd.DataFrame(s17,columns=['ID','Name','Price','Rate','Quantity','id_c'])
df.head()
```
